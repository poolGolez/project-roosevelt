```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory

RANDOM_SEED = 144
INPUT_DIR = '/kaggle/input/apples-or-tomatoes-image-classification/train'
```

```python
train_data = image_dataset_from_directory(INPUT_DIR, seed=RANDOM_SEED)
```

```python
train_data.class_names
```

```python
batch  = next(iter(train_data))
_, axs = plt.subplots(4, 8, figsize=(20, 10))
for (index, data) in enumerate(zip(batch[0], batch[1])):
    ax = axs[index // 8, index % 8]
    image = data[0]
    label = train_data.class_names[data[1].numpy()]
    ax.imshow(image/255.)
    ax.set_title(label)
    ax.axis('off')
    
```
